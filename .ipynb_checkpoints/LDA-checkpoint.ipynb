# Estimating Rebalancing Needs
One of the most interesting and important problems with bike share programs is that of rebalancing the system when asymmetric use occurs. Usually, this type of asymmetry is the result of some external factor like elevation or simple one-way rides. In the analysis below I will attempt to develop a way to spot asymmetrical ridership in the bay area bike share program. I will not be attempting the actual rebalancing optimization, but rather suggesting the number of bikes that need to be added to a given station and where they can come from (How they are added is a whole different can of worms).

All of my analysis and reccomendations will be focused in San Francisco. The other cities in the south bay have low ridership numbers and are subsequently not that interesting.

### Ok...It's All Commuters. So What?

The great thing about commutes is that they are generally symmetric; someone will travel from station A to station B for the day and reverse that exact route on the way home. This is enormously helpful when it comes to rebalancing. However, there will still be stations that end up unbalanced for various reasons. The goal of the rest of this analysis is to spot those unbalanced stations and reccommend a rebalancing action so that those stations can be used.

# Calculating Net Station Flow
Net station flow is defined as (Rides_Ending - Rides_Starting) at a single station in a time period. If net station flow is negative, more bikes are being ridden from the station than to it in that time period. With the work commute in mind, the relevant time periods are AM and PM.  This will be the statistic that we focus on to determine whether or not a station will need rebalancing.

It is important to note that we do not have inventory data. We can know the change in inventory at a station in a time period, but we do not know initial invetory for that station, so we cannot know ending inventory. This makes suggesting inventory movement problematic. We can still suggest the amount of inventory to add/remove in order to bring the station back to initial inventory. (The system is probably already rebalancing, so we can look at this a tune to the exisitng rebalancing model. This will be addressed later.)
### DOW and TOD Aggregation

```python
####Rides Started ToD####
dftemps_t = df0.groupby(["Start Station", 'DoWS', df0['Start Date'].dt.date, 'ToDS']).count()['Duration'] #rides started by station per day
####Rides Ended ToD####
dftempe_t =df0.groupby(["End Station", 'DoWE', df0['End Date'].dt.date, 'ToDE']).count()['Duration']#rides ended by station per day
```

Above groups and counts the initial dataframe on station, date and time of day for both rides started and ended. This counts all rides started at a given station for the day of week, date and time of day (see example structure below). The goal will be to subtract rides starting at a give station from the rides ending to get a net bike flow for a given time period. This net flow will be used to approximate the change in inventory at the station.

```python
dftemps_t.head()
```

```python
dftempe_t.index.levels[1].name='DoWS'
dftempe_t.index.levels[0].name='Start Station'
dftempe_t.index.levels[2].name = 'Start Date'
dftempe_t.index.levels[3].name = 'ToDS'
#name indexes the same for both dataframes so that they may be subtracted
```

### Sample Size

Because there are only about 25 weeks in the dataset, we end up with only 25 data points for every mean net flow statistic. This is a pretty small sample, so we can't be certain that the means are representative of the population, i.e. the actual mean net flow for a station. 

### Bootstrapping
A way around this is to resample the data randomly and compute a confidence interval from the resampled data. Below we calculate the 90% confidence interval for the sample mean net flow with a resample size of 100.

```python
import scipy as sp
import scipy.stats as stats
import math
```

```python
def bootstrap_tod():
    grouped_sample = dftempe_t.sub(dftemps_t).fillna(0).groupby(['Start Station', 'DoWS', 'ToDS'])#subtract dataframes, group
    sample = grouped_sample.agg([np.mean]) #calculate sample means
    sample['Conf_L'] = 'Nan'
    sample['Conf_H'] = 'Nan'

    n=100 #bootstrap sample size
    strap = {}
    for i in range(n):
        experiment = dftempe_t.sub(dftemps_t).fillna(0).sample(n=11649, replace=True) #sample w/ same dimensions + replacement
        grouped = experiment.groupby(['Start Station', 'DoWS', 'ToDS'])
        strap[i] = grouped.agg([np.mean]) #calc means of bootstrap samples

    for index in sample.index:
        arr = []
        for i in range(n):
            try:
                arr.append(strap[i].loc[index].values[0]-sample.loc[index].values[0]) #find deviations from sample mean, handle missing station data 
            except:
                arr.append(False)
        arr.sort()
        sample.set_value(index,'Conf_L', arr[5])
        sample.set_value(index,'Conf_H', arr[95])#assign approximate 90% confidence interval to sample mean dataframe
    return sample
```

```python
station_locations.index.name = 'Start Station'
sample_tod=bootstrap_tod().join(station_locations[['landmark', 'dockcount']])
```

Below you can see the mean net flow in the AM and PM hours, respectively, by day of week. You will see that at some stations there is symmetric flow (2nd at South Park) and asymmetric flow at others (2nd at Folsom) during the week days. It would be ideal to rebalance the asymmetric stations at midday as far as user experience is concerned so that no station is unbalanced all day from the AM rides. Just repeating the above steps modulo the time of day.

```python
####Rides Started DAY####
dftemps_d = df0.groupby(["Start Station", 'DoWS',df0['Start Date'].dt.date]).count()['Duration'] #rides started by station per day
```

```python
####Rides Ended DAY####
dftempe_d =df0.groupby(["End Station", 'DoWE', df0['End Date'].dt.date]).count()['Duration'] #rides ended by station per day
```

```python
dftempe_d.index.levels[1].name='DoWS'
dftempe_d.index.levels[0].name='Start Station'
dftempe_d.index.levels[2].name = 'Start Date'
```

```python
def bootstrap_dow():
    grouped_sample = dftempe_d.sub(dftemps_d).fillna(0).groupby(['Start Station', 'DoWS'])
    sample = grouped_sample.agg([np.mean])
    sample['Conf_L'] = 'Nan'
    sample['Conf_H'] = 'Nan'

    n=100 #samples
    strap = {}
    for i in range(n):
        experiment = dftempe_d.sub(dftemps_d).fillna(0).sample(n=11649, replace=True)
        grouped = experiment.groupby(['Start Station', 'DoWS'])
        strap[i] = grouped.agg([np.mean])

    for index in sample.index:
        arr = []
        for i in range(n):
            try:
                arr.append(strap[i].loc[index].values[0]-sample.loc[index].values[0])
            except:
                arr.append(False)
        arr.sort()
        sample.set_value(index,'Conf_L', arr[5])
        sample.set_value(index,'Conf_H', arr[95])
    return sample
```

```python
station_locations.index.name = 'Start Station'
sample_dow=bootstrap_dow().join(station_locations[['landmark', 'dockcount']])
```

Below is the net mean flow by day of week (with rebalance recommendations). This can be interpreted as the number of bikes that a station will on average accumulate or lose over the course of a day. Assuming that rebalacing will happen once daily at the end of the day, we can return the stations to their initial inventory levels by cancelling out the net change in inventory.

### Deciding to Rebalance
For this decision we can leverage that handy confidence interval calculated above. The logic here is that if the null hypothesis (i.e. 0 net mean flow) is contained within the confidence interval, the net mean flow is not statistically significant and we have failed to reject the null hypothesis. If, however, the null hypothesis is outside the interval, we can say that the net mean flow is statistically significant (to our 90% CL) and we should expect to rebalance the station.

Once we have decided if we should rebalance, we need to decide how. If the net mean flow is negative, we should add bikes, if positive, we should remove bikes. How many bikes? Some number of bikes in the CI, i.e. between 'Adjust Low End' and '	Adjust High End' in the output below.

**The recommendations are more easily viewed in the tool included with this analysis. The tool also makes recommendations for AM/PM rebalancing. Note becuase the intervals have been recalculated, the values will not be the same as in the tool, but the recommendations should be valid.**

### Detecting Existing Rebalancing

As mentioned before, the system is already using some sort of rebalancing. I first noticed this when looking at San Francisco Caltrain (Townsend at 4th) station. There are only 19 docks at this station yet there were many more rides flowing through the station according to the data. This meant that there were bikes being added to the station.

To detect rebalancing, we have to look at breaks in station continuity for a given bike. That is, for a given bike if we arrange all rides in chronological order and the start station does not equal the end station of the previous ride, we know that the bike was moved from the end station to the start station! The problem with using ride data to pick out rebalances is that we cannot know exactly when the move ocurred. The bike could have been moved at any point between the end time and start time of the rides in question. This could potentially cause problems when aggregating rebalace moves on start/end timestamps of rides.

Ideally, we would add and subtract these bike movements grouped by day of week to/from the mean net flow in order to obtain "normalized" net mean flow values. Then resample that data to obtain a CI and base rebalancing decisions on that. Unfortunately time is limiting the extent of this analysis.

I have included sample mean bike movements by station per day below to give an idea of how rebalancing is ocurring.

```python
# import json
# with open('./demand_dow.json', 'w') as outfile:
#     json.dump(ad, outfile)
# with open('./demand_tod.json', 'w') as outfile:
#     json.dump(ad, outfile)
```

```python
bikes = np.sort(df0['Bike #'].unique())
rebcount_all = {}
for bike in bikes: 
    rebcheck = df0.loc[df0[df0['Bike #'] == bike]['Start Date'].sort_values().index]
    to_ = rebcheck[rebcheck['Start Terminal'] != rebcheck['End Terminal'].shift(1)][['Start Station', 'Start Date']]
    from_ = rebcheck[rebcheck['End Terminal'] != rebcheck['Start Terminal'].shift(-1)][['End Station', 'End Date']]
    from_.reset_index(inplace=True)
    to_=to_.iloc[1:]
    to_.reset_index(inplace=True)
    reb = from_.join(to_ , how='left', rsuffix='_to', lsuffix='_from')
    rebcount_all[bike] = reb
```

```python
rebtot = rebcount_all[569]
for bike in bikes:
    if bike != 569:
        rebtot = rebtot.append(rebcount_all[bike], ignore_index=True)
```
